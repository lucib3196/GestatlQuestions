```python
import re
pattern = r'</pl-[a-z\-]+>'
unique_tags = set()
filtered = df[~df["isAdaptive"].isna()]
for text in filtered["question.html"]:
    if isinstance(text, str):
        result = re.findall(pattern, text)
        # if "</pl-mca-adaptive>" in result:
        #     print(text)
        unique_tags.update(result)
        
print(unique_tags)
```
